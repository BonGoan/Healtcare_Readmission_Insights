```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort the DataFrame based on Test_recall in descending order
df_metrics_sorted = df_metrics.sort_values(by="Test_recall", ascending=False)

# Create a figure with two subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(16, 6))  # figsize controls the overall size

# Plot 1: Test Recall
sns.barplot(df_metrics_sorted, y="Model", x="Test_recall", ax=axes[0])  # Use the first subplot
axes[0].set_title('Test Recall')
axes[0].set_xlabel('Test Recall')
axes[0].set_ylabel('Model')

# Add values to the Test Recall bars
for index, value in enumerate(df_metrics_sorted['Test_recall']):
    # Set vertical position
    if index < 2:  # For the top two models
        vertical_offset = 0.03  # Adjust upward for top two
    else:
        vertical_offset = -0.03  # Adjust downward for the rest
    axes[0].text(value, index + vertical_offset, f"{value:.2f}", va='center')  # Add text to each bar

# Plot 2: Test F1 Score
sns.barplot(df_metrics_sorted, y="Model", x="Test_f1", ax=axes[1])  # Use the second subplot
axes[1].set_title('Test F1 Score')
axes[1].set_xlabel('Test F1')
axes[1].set_ylabel('')

# Add values to the Test F1 Score bars
for index, value in enumerate(df_metrics_sorted['Test_f1']):
    # Set vertical position
    if index < 2:  # For the top two models
        vertical_offset = 0.03  # Adjust upward for top two
    else:
        vertical_offset = -0.03  # Adjust downward for the rest
    axes[1].text(value, index + vertical_offset, f"{value:.2f}", va='center')  # Add text to each bar

# Adjust layout to prevent overlap
plt.tight_layout()

# Define the path to save the combined figure
save_path = '../03_figures/combined_recall_f1_barplot.png'

# Save the plot to the specified folder
plt.savefig(save_path, bbox_inches='tight', dpi=300)

# Show the merged plot
plt.show()
```
